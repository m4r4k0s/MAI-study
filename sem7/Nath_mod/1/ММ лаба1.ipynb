```python
import matplotlib.pyplot as plt
import scipy.integrate
import numpy as np


C=0
f = lambda x: 1/np.sqrt(2*C- x**2 - (2*x**2+4)/np.sqrt(x**2 + 1))
C0 = lambda x: x**2/2 +(x**2 + 2)/np.sqrt(x**2 + 1)
y = []
x = np.arange(0.1,1.74,0.02)
for a in x:
    C=C0(a)
    y.append(2*scipy.integrate.quad(f, -a, a)[0])
    print(a,y[-1])
fig, ax = plt.subplots()
print(y)
ax.plot(x, y)
ax.grid()
ax.set_xlabel('a')
ax.set_ylabel('T')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math


class color:
    def __init__(self):
        self.colors =['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
        self.i=0
    
    def next(self):
        if self.i<len(self.colors)-1:
            self.i+=1
            return(self.colors[self.i-1])
        else:
            self.i=0
            return(self.colors[len(self.colors)-1])


def di(args):
    try:
        return math.sqrt(2*args[0]-args[1]**2-(2*args[1]**2 + 4)/math.sqrt(1+args[1]**2))
    except:
        return 0 


def main():
    cl = color()
    C=0
    while C < 6:
        x= np.arange(-5,5.01,0.01)
        y1 = [float(di([C,x[i]])) for i in range(len(x))]
        y2 = [float(-di([C,x[i]])) for i in range(len(x))]
        lc = cl.next()
        plt.plot(x,y1, color = lc)
        plt.plot(x,y2, color = lc)
        C+=0.5
    ax = plt.gca()
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.show()

if __name__ == "__main__":
    main()
```
