```python
def error(U):
    x1, y1 = np.mgrid[0:np.pi/2:100j, 0:np.pi/2:100j]
    z1 = np.exp(-y1)*np.cos(x1)*np.cos(y1)
    X = np.linspace(0, np.pi/2, 100)
    Y = []
    for t in range(100):
        Y.append(np.max(np.abs(U[t]-z1[t])))
    print("средняя ошибка = {}".format(np.mean(Y)))
    plt.plot(X, Y, label = 'error')
    plt.xlabel('График ошибки')
    plt.legend()
    plt.show()
```

```python
def graf_in_time(u, f, y, tau):
    X = np.linspace(0, np.pi/2, 100)
    z1 = f(y*tau, X)
    plt.plot(X, z1, label = 'точно')
    plt.plot(X, u[y], label = 'численно')
    plt.xlabel('График функций')
    plt.legend()
    plt.show()
```

```python
f = lambda x,y: np.exp(-y)*np.cos(x)*np.cos(y)
z2 = np.array(pd.read_csv('res.csv'), dtype=np.double)
tau=(np.pi/2)/100
error(z2)
graf_in_time(z2,f, 70, tau)
```
