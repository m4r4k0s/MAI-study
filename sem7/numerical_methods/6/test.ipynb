```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x,t: np.exp(-t)*np.sin(x)
k = 1000
n = 100
l = np.pi
t = 1
tau = t/(k-1)
h = l/(n-1)
```

```python
def error(U, fR):
    t1, x1 = np.mgrid[0:1:1000j, 0:np.pi:100j]
    z1 = np.exp(-t1)*np.sin(x1)
    X = np.linspace(0, 1, 1000)
    Y = []
    for t in range(1000):
        Y.append(np.max(np.abs(U[t]-z1[t])))
    print("средняя ошибка = {}".format(np.mean(Y)))
    plt.plot(X, Y, label = 'error')
    plt.xlabel('График ошибки')
    plt.legend()
    plt.show()
```

```python
def graf_in_time(u,f,t, tau):
    X = np.linspace(0, np.pi, 100)
    z1 = f(X,t*tau)
    plt.plot(X, z1, label = 'точно')
    plt.plot(X, u[t], label = 'численно')
    plt.xlabel('График функций')
    plt.legend()
    plt.show()
```

```python
z2 = np.array(pd.read_csv('res.csv'), dtype=np.double)
error(z2, f)
graf_in_time(z2,f,999, tau)
```
